```
# 1
import pandas as pd

# Load the regular season and playoff data (semicolon-separated and encoding handled)
df = pd.read_csv('/content/DS2002F24/homework/acc_players-2324F.csv', sep=',', encoding='latin1', skiprows=[0]) # skipping first row because they are not useful headings

display(df.head(5)) # displaying first five rows
```

```
# 2
# total points scored by all players
total = df['PTS'].sum()
print(total)
```

```
# player with most minutes
df['Player'].iloc[df['MP'].idxmax()]
```

```
# top 5 players in terms of total rebounds
df.nlargest(5, 'TRB')['Player']
```

```
# 3
# creating new df containing only players who played more than 500 min
df_new = df[df['MP'] > 500]
df_new.head()
```

```
# determining player with the highest total assists
df_new['Player'].iloc[df_new['AST'].idxmax()]
```

```
# top 3 assist leads in league (using unfiltered df)
df.nlargest(3, 'AST')['Player']
```

```
# top 3 shot blockers (using unfiltered df)
df.nlargest(3, 'BLK')['Player']
```

```
# 4
# total points scored by each school
df_school = df.groupby('School')[['PTS']].sum().reset_index()
df_school
```

```
# total assists for each school
df_ast = df.groupby('School')[['AST']].sum().reset_index()
df_ast
```

```
# sort schools by total pts scored and display top 3 schools
df_school.nlargest(3, 'PTS')['School']
```
